Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
Q1 = data['average_rating'].quantile(0.25)
Q3 = data['average_rating'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

# Menampilkan data yang merupakan outlier
data_clean = data[(data['average_rating'] >= batas_bawah) & (data['average_rating'] <= batas_atas)]
len(data_clean)
```

```python
data_clean.loc[:, 'actual_price'] = pd.to_numeric(data_clean['actual_price'], errors='coerce')
data_clean.loc[:, 'selling_price'] = pd.to_numeric(data_clean['selling_price'], errors='coerce')

# 2. Tentukan kolom numerik yang ingin diisi nilai kosongnya
numeric_cols = ['actual_price', 'selling_price', 'average_rating']

# 3. Inisialisasi imputer
imputer = SimpleImputer(strategy='mean')

# 4. PERBAIKAN: Gunakan data_clean, JANGAN data asli agar sinkron dengan tipe data baru
data_clean[numeric_cols] = imputer.fit_transform(data_clean[numeric_cols])
data[numeric_cols].isna().sum()
```

```python
X = data_clean[numeric_cols].drop('average_rating', axis=1)
y = data_clean['average_rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
```

```python
scaler = MinMaxScaler()


X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
